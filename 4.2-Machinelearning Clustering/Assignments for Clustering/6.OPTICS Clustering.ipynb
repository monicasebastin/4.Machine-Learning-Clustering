```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
dataset=pd.read_csv("Mall_Customers.csv")
dataset
```

```python
x=dataset.iloc[:,3:5]
x
```

```python
from sklearn.neighbors import NearestNeighbors
NN=NearestNeighbors(n_neighbors=10)
neighbors=NN.fit(x)
distances,indices=neighbors.kneighbors(x)

plt.figure(figsize=(5,2))
plt.plot(np.sort(distances[:,9]))
plt.xlabel="values"
plt.ylabel="epsilon"
```

```python
E=20
from sklearn.cluster import OPTICS,cluster_optics_dbscan
OP=OPTICS(min_samples=10)
OP.fit(x)
pred=cluster_optics_dbscan(
    reachability=OP.reachability_,
    core_distances=OP.core_distances_,
    ordering=OP.ordering_,
    eps=E,)
```

```python
supervised=dataset
supervised['Cluster_group']=pred
supervised
```

```python
supervised.to_csv("OPTICS Cluster.csv",index=False)
```

```python
import seaborn as sns
facet=sns.lmplot(data=supervised,x=supervised.columns[3],y=supervised.columns[4],hue=supervised.columns[5],fit_reg=False,legend_out=True)
```

```python
cluster=len(np.unique(pred))
from sklearn.metrics import silhouette_score
score=silhouette_score(x,pred,sample_size=1000,random_state=42)
print(f"Epsilon={E},cluster size={cluster},Score={score}")
```
