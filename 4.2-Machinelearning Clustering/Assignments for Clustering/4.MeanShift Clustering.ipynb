```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
dataset=pd.read_csv("Mall_Customers.csv")
dataset
```

```python
x=dataset.iloc[:,3:5].values
x
```

```python
from sklearn.cluster import estimate_bandwidth, MeanShift

bandwidth = estimate_bandwidth(x, quantile=0.2, n_samples=500)
ms = MeanShift(bandwidth=bandwidth)
ms.fit(x)
labels = ms.labels_
```

```python
labels = ms.labels_
cluster_centers = ms.cluster_centers_
```

```python
supervised = pd.DataFrame(x)
supervised['Cluster_group'] = labels
supervised
```

```python
supervised.to_csv("MeanShift.csv",index=False)
```

```python
import seaborn as sns
sns.scatterplot(
    data=supervised,
    x=supervised.columns[0],
    y=supervised.columns[1],
    hue='Cluster_group',
    palette='Set1',
    s=50
)

# Plot cluster centers
plt.scatter(
    cluster_centers[:, 0],
    cluster_centers[:, 1],
    c='black',
    s=150,
    marker='x',
    label='Centers'
)

plt.title("MeanShift Clustering with Cluster Centers")
plt.legend()
plt.show()
```

```python
cluster=len(np.unique(labels))
from sklearn.metrics import silhouette_score
score=silhouette_score(x,labels,sample_size=1000,random_state=42)
print(f"cluster size={cluster},Score={score}")
```
